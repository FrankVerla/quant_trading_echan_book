>## Setting up your business

As a retail trader, you have two main choices:

1. open a retail brokerage account
2. join a proprietary trading firm

*In a nutshell, retail borkerages give you complete freedom and better capital protection but smaller leverage, while proprietary trading firms give you less freedom and less capital protection but much high leverage.*

### Open a retail brokerage account

No one will question your strategy, and no one will guide you in your trading. Your leverage is limited by Securities and Exchange Commission (SEC) Regulation T (2x your equity if hold overnight).


### Join a prop trading firm

Here are some examples: Bright Trading, ECHOtrade, etc.

Become a member of a prop trading firm, and pass the National Association of Securities Dealers (NASD) Series 7 examination to qualify you as a registered representative of a brokerage.

In terms of liability, your loss is limited to your initial investment. Actually, liability is also limited if you form an S corporation of limited liability company [LLC] and open an account through this entity at a retail brokerage.

Another consideration that applies to those of you who have discovered some unique, highly profitable strategies. In this situation, you may prefer to open a retail trading account, because if you trade through a proprietary account, your proprietary trading firm is going to find out about your highly profitable strategy and may "piggback" on your strategy with a lot of its own capital.

### Lastly, consider tax

## How to choose a brokerage or a prop trading firm

1. commission rate
2. the speed of execution of your brokerage
3. their access of the so-called "dark-pool" liquidity
4. the range of products you can trade
5. application programming interface (API) of a trading platform
6. availability of paper trading accounts / simulator account
    - quotes from the past are displayed as if they were real-time quotes, and an automated trading program can trade against these quotes at any time of the day in order to debug the program
7. reputation and financial strenght of prop trading firm
8. you can actually open a brokerage AND a prop trading account
    - as long as this fact is fully disclosed to the prop trading and to NASD as "outside business activities" and prior permission obtained.
    
**Make sure your trading account/system have:**

1. Relatively low commissions
2. Trade a good variety of financial instruments
3. Access to deep pool of liquidity
4. Most Importantly, API for real-time data retrieval and order transmission

## Build physical infrastructure

In the start-up phase of your business, the physical infrastructure can be light and simple

- A good personal computer
- A high-speed internet connection
- An uninterruptible power supply (UPS) 
    - so that your computer does not accidentally shut down.
- Some professional real-time newsfeed
    - Thomson Reuters
    - Dow Jones
    - Bloomberg
- Multiple monitors to hook up to the same computer
    - so that you have extended screen space to monitor all the different trading applications and portfolios.
- Serve hosting or collocation
